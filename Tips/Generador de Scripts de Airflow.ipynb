# AIRFLOW SCRIPT GENERATOR

```python
import random
from datetime import datetime
from textwrap import dedent
import os

def generar_script_airflow(posibles_variables, lineas_offset):
    # Plantilla base del script
    script_base = dedent("""
    from airflow import DAG
    from airflow.operators.python_operator import PythonOperator
    from airflow.models import Variable
    from datetime import datetime

    # Definir DAG
    default_args = {
        'owner': 'airflow',
        'depends_on_past': False,
        'start_date': datetime(2023, 8, 9),
        'retries': 1,
    }

    dag = DAG(
        'conexion_varia_{index}',
        default_args=default_args,
        description='Un DAG que utiliza una variedad de variables de Airflow',
        schedule_interval=None,
    )

    # Función para realizar la tarea
    def realizar_tarea(**kwargs):
        # Capturar variables
    """)

    # Seleccionar un número aleatorio de variables de la lista de posibles variables
    numero_variables = random.randint(1, lineas_offset)
    variables_seleccionadas = random.sample(posibles_variables, numero_variables)

    # Generar script con las variables seleccionadas y valores aleatorios
    script_variables = ""
    for var in variables_seleccionadas:
        valor_random = random.choice(posibles_variables)
        valor_random_default = random.choice(posibles_variables)
        script_variables += f"    {var} = Variable.get('{valor_random}', default_var='{valor_random_default}')\n"

    # Seleccionar un número aleatorio de variables de la lista de posibles variables
    numero_variables = random.randint(1, lineas_offset)
    variables_seleccionadas = random.sample(posibles_variables, numero_variables)
    
    script_variables_finales = ""
    for var in variables_seleccionadas:
        valor_random = random.choice(posibles_variables)
        valor_random_default = random.choice(posibles_variables)
        script_variables_finales += f"    {var} = Variable.get('{valor_random}', default_var='{valor_random_default}')\n"

    
    script_tareas = dedent(f"""
    # Asignar variables a funciones
{script_variables}

    print("Variables cargadas exitosamente.")

    # Asignar nuevas variables a funciones
{script_variables_finales}
    """)

    # Definir la tarea de Python
    script_tareas += dedent("""
        realizar_tarea_task = PythonOperator(
            task_id='realizar_tarea',
            provide_context=True,
            python_callable=realizar_tarea,
            dag=dag,
        )

        realizar_tarea_task
    """)

    # Generar el script completo
    script_completo = script_base + script_tareas.format(index=lineas_offset)
    return script_completo
```

# PYTHON FILES GENERATORS

```python
def guardar_scripts(cantidad, posibles_variables, numero_variables, carpeta_salida):
    # Crear la carpeta de salida si no existe
    if not os.path.exists(carpeta_salida):
        os.makedirs(carpeta_salida)
    
    for i in range(cantidad):
        script = generar_script_airflow(posibles_variables, numero_variables)
        nombre_archivo = f"{carpeta_salida}/script_{i + 1}.py"
        
        with open(nombre_archivo, 'w') as archivo:
            archivo.write(script)
        print(f"Script guardado en: {nombre_archivo}")
        
cantidad_scripts = 13  # Número de scripts a generar
numero_variables = 7  # Número de variables por script
carpeta_salida = "/notebooks/scripts_airflow"

guardar_scripts(cantidad_scripts, posibles_variables, numero_variables, carpeta_salida)
```

# READ AIRFLOW VARIABLES FROM PYTHON SCRIPTS

```python
import os
import random
import re
import pandas as pd
# Configurar pandas para mostrar el contenido completo de las celdas
pd.set_option('display.max_colwidth', None)

def buscar_variable_get(directorio, posibles_variables):
    datos = []

    # Expresión regular para capturar la asignación de la variable y los valores en Variable.get()
    patron_variable = re.compile(
        r"(\w+)\s*=\s*Variable\.get\(\s*['\"](.+?)['\"](?:,\s*default_var=(['\"]?.+?['\"]?)|,\s*([\d\w]+))?\s*\)"
    )

    # Recorrer todos los archivos en el directorio
    for nombre_archivo in os.listdir(directorio):
        if nombre_archivo.endswith('.py'):
            ruta_archivo = os.path.join(directorio, nombre_archivo)

            with open(ruta_archivo, 'r') as archivo:
                lineas = archivo.readlines()

                # Buscar las líneas que contienen 'Variable.get'
                for numero_linea, linea in enumerate(lineas, start=1):
                    # Encuentra todas las coincidencias en una línea
                    matches = patron_variable.findall(linea)

                    if matches:
                        for match in matches:
                            variable_name = match[0]  # Nombre de la variable en el script
                            variable_value = match[1]  # Valor usado en Variable.get()
                            default_variable_value = match[2] if match[2] else match[3]  # Valor por defecto o vacío
                            valor_reemplazo = random.choice(posibles_variables)

                            datos.append({
                                'script': nombre_archivo,
                                'line': numero_linea,
                                'content': f"{variable_name} = Variable.get('{variable_value}', default_var={default_variable_value})",
                                'variable_name': variable_name,
                                'variable_value': variable_value,
                                'default_variable_value': default_variable_value if default_variable_value and default_variable_value != "''" else '',
                                'replace': valor_reemplazo
                            })

    # Crear un DataFrame de pandas
    df = pd.DataFrame(datos)
    df.index += 1  # Ajustar el índice para que comience en 1
    return df


# Ejemplo de uso
directorio = '/notebooks/scripts_airflow/'
df_resultado = buscar_variable_get(directorio, posibles_variables)
df_resultado
```

# GENERATE NEW PYTHON SCRIPTS FIXED

```python
import os
import pandas as pd

def reemplazar_y_guardar_scripts(df, directorio_original, directorio_salida):
    if not os.path.exists(directorio_salida):
        os.makedirs(directorio_salida)

    # Agrupar por nombre del script
    for nombre_script, grupo in df.groupby('script'):
        ruta_original = os.path.join(directorio_original, nombre_script)
        ruta_salida = os.path.join(directorio_salida, nombre_script)
        
        # Leer el contenido original del script
        with open(ruta_original, 'r') as archivo:
            lineas = archivo.readlines()

        # Modificar las líneas según el DataFrame
        lineas_modificadas = lineas[:]
        for _, fila in grupo.iterrows():
            numero_linea = fila['line'] - 1  # Ajustar al índice basado en 0
            valor_reemplazo = fila['replace']
            
            # Reemplazar el valor después del igual
            if ' = ' in lineas_modificadas[numero_linea]:
                partes = lineas_modificadas[numero_linea].split('=')
                # Limpiar espacios y comillas para el nuevo valor
                nuevo_valor = f'"{valor_reemplazo.strip()}"'
                lineas_modificadas[numero_linea] = f"{partes[0].strip()} = {nuevo_valor}\n"

        # Guardar el script modificado
        with open(ruta_salida, 'w') as archivo:
            archivo.writelines(lineas_modificadas)
        print(f"Script modificado guardado en: {ruta_salida}")

# Ejemplo de uso
directorio_original = '/notebooks/scripts_airflow'  # Directorio con los scripts originales
directorio_salida = '/notebooks/scripts_airflow_fixed'  # Directorio para los scripts modificados

reemplazar_y_guardar_scripts(df_resultado, directorio_original, directorio_salida)
```

# GENERATE NEW PYTHON SCRIPTS FIXED WITH VARIABLES ON SAME LINE

```python
import os

def reemplazar_y_guardar_scripts_multivariables(df, directorio_original, directorio_salida):
    if not os.path.exists(directorio_salida):
        os.makedirs(directorio_salida)

    # Agrupar por nombre del script
    for nombre_script, grupo in df.groupby('script'):
        ruta_original = os.path.join(directorio_original, nombre_script)
        ruta_salida = os.path.join(directorio_salida, nombre_script)
        
        # Leer el contenido original del script
        with open(ruta_original, 'r') as archivo:
            lineas = archivo.readlines()

        # Crear un diccionario para actualizar líneas basadas en el DataFrame
        lineas_actualizadas = {}

        for _, fila in grupo.iterrows():
            numero_linea = fila['line'] - 1  # Ajustar al índice basado en 0
            variable_name = fila['variable_name']
            valor_reemplazo = fila['replace']
            
            # Crear una línea modificada con el nuevo valor
            nueva_linea = f"{variable_name} = \"{valor_reemplazo.strip()}\"\n"
            
            if numero_linea in lineas_actualizadas:
                # Si ya hemos añadido una línea para esta línea original, agregar el nuevo reemplazo
                lineas_actualizadas[numero_linea] += nueva_linea
            else:
                # Añadir una nueva entrada para la línea
                lineas_actualizadas[numero_linea] = nueva_linea

        lineas_modificadas = []

        # Crear la lista de líneas finales
        
        for i, linea in enumerate(lineas):
            if i in lineas_actualizadas:
                # Añadir las líneas actualizadas para las líneas que tienen reemplazos
                #lineas_modificadas.append(lineas_actualizadas[i])
                pass
            else:
                # Conservar la línea original si no se actualiza
                lineas_modificadas.append(linea)

        
        # Añadir líneas adicionales si alguna línea modificada tiene múltiples partes
        for key in sorted(lineas_actualizadas.keys()):
            if len(lineas_actualizadas[key].split('\n')) > 1:
                # Insertar líneas adicionales en el lugar correcto
                partes = lineas_actualizadas[key].split('\n')
                for parte in partes:
                    if parte:
                        lineas_modificadas.insert(key + 1, parte + '\n')
                        key += 1

        # Guardar el script modificado
        with open(ruta_salida, 'w') as archivo:
            archivo.writelines(lineas_modificadas)
        print(f"Script modificado guardado en: {ruta_salida}")

# Ejemplo de uso
directorio_original = '/notebooks/scripts_airflow'  # Directorio con los scripts originales
directorio_salida = '/notebooks/scripts_airflow_fixed_multivariables'  # Directorio para los scripts modificados

reemplazar_y_guardar_scripts_multivariables(df_resultado, directorio_original, directorio_salida)
```
